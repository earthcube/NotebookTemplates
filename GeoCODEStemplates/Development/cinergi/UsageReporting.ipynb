# Usage reporting

## Notes

This notebook is a start for generating Usage reports from EarthCube applications that would be posted to an annotation server. Intention is that the content acquired should be compatible with the ISO19115 MD_Usage element, as well as W3C Prov or [W3C Annotation](https://www.w3.org/TR/annotation-model/)  information models.

### Usage Annotation:

* annotationID: unique identifier for this annotation, probably autogenerated UUID

* subjectID: a URI that identifies the resource this annotation is about.  Should be a resolvable identifier that enables linking to the documented resource.

* context:  some way of identifying the resource used, and ideally the software environment of application. Free text to start, ideally some machine-actionable encoding could be provided.

* specificUsage (annotationNote):  free text, describe how data were used

* usageDateTime(annotationDateTime): time stamp, defaults to time that report is submitted to annotation server, also allow user-specified time when the usage was attempted (must be previous to the actual report time)

* userDeterminedLimitations (optional):  free text. Describe any problems discovered in the data content or its usage with specific applications

* source contact information (optional): responsible party for report; select appropriate role on CI_REsponsibleParty



### Comment Annotation:

* AnnotationID: unique identifier for this annotation, probably autogenerated UUID

* subjectID: URI for resource that is subject of the annotation; could be dataset, metadata record, another annotation.  Should be a resolvable identifier that enables linking to the documented resource.

* annotationNote: Free text description of annotation comment

* annotationDateTime: time stamp, defaults to time that report is submitted to annotation server

* source contact information (optional): responsible party for report; select appropriate role on CI_REsponsibleParty
